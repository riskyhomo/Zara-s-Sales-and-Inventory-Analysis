#### Functions

```python
def merge_dataframe (objs, key):
    """Merge DataFrames, given the dfs (objs) and the key (where df will merge)"""
    return reduce(lambda left,right: pd.merge(left,right,on=key, how='outer'), objs)
```

```python
def dataframe_clean (df):
    """Cleans the dataframes from wbdata (world bank API library).
    Removes all the geographical and income groups from the dataframe. """
    
    df = df.reset_index(drop=False)
    row = df.loc[df['country'] == 'World'].index.item()
    df = df[(row+1):]
    df = df.reset_index(drop=True)
    return df
```

## Sales & Stock by day

Since this dataset was released in April 2019 and the challenge related with with was to predict the products that would have the most sales, we can infer that this data set was recent. 

As we can see below, there was a pick in sales in the day 15 and in the day 79. 
If our conclusions are right, day 15 corresponds to the end of January, and day 79 corresponds to the end of March. This two months are also months of sales and price reductions. Since March is usually the last month of sales it is expectable that the highest pick is during this period.

## Revenue & Inventory Analysis

Sales and Inventory and the most important figures in retail. As sales are the way the company makes money, it is also important to look to the inventory that is stocked. Inventory represents one of the most important assets of a business because the turnover of inventory represents one of the primary sources of revenue generation and subsequent earnings for the company's shareholders.

```python
products_final['revenue'] = products_final['price']*products_final['sales']

products_final['curr_liability'] = products_final['price']*products_final['stock']

revenues_day = pd.pivot_table(products_final, 
                              values=['revenue', 'curr_liability'], 
                              index='date_number',
                              aggfunc=np.sum)
revenues_day
```

```python
revenues_day.sort_values(by='revenue')
```

### Inventory Turnover by Product

```python
invent_turnover_table = pd.pivot_table(products_final, 
                                 values=['sales', 'stock', 'revenue', 'curr_liability', 'price'], 
                                 index='product_id',aggfunc={'sales':'sum', 
                                                             'stock':'mean', 
                                                             'revenue': 'sum',
                                                            'curr_liability':'sum',
                                                            'price':'mean'})
```

#### Absolute sales vs stock

```python
invent_turnover_table['inventory_turnover'] = invent_turnover_table['sales']/invent_turnover_table['stock']

invent_turnover_table.sort_values(by='inventory_turnover', ascending = True).head(5)
```

There are several products with no sales during this 91 days period. For this analysis it is important to understand which are this products and what they represent in the total numbers. We are going to drop this products so it doesn't impact the next calculations. 

```python
n_products = len(invent_turnover_table)

no_revenue = invent_turnover_table[invent_turnover_table['sales']==0]
no_revenue_products_list = list(no_revenue.index.to_numpy())
no_revenue_products = len(no_revenue_products_list)

print (f"There around {n_products} products. From this, {no_revenue_products} didn't have sales. \
This represents around {round(no_revenue_products/n_products*100)}%.")
```

```python
print (no_revenue_products_list)
```

```python
avg_inventory_turnover = round(invent_turnover_table['inventory_turnover'].mean(),2)

print (f'The average inventory turnover is {avg_inventory_turnover} for this 91 days period.')
```

Accordingly to [CSI](https://csimarket.com/Industry/industry_Efficiency.php?ind=1301), the retail appareal industry has an average inventory turnover of 8.22. Zara's inventory turnover is lower than this, which is a good indicator.

### Days sales of Inventory
This ratio indicates how long it takes to clear the inventory. 

```python
sales_days = len(list(products_final.date_number.unique()))
dsi = round(sales_days/avg_inventory_turnover,2)
print (f'It takes, on average, {dsi} days for Zara to clear its inventory.')
```

```python
invent_turnover_table['dsi'] = sales_days/invent_turnover_table['inventory_turnover']
invent_turnover_table
```

```python
invent_turnover_table.dropna(inplace=True)
invent_turnover_table = invent_turnover_table[invent_turnover_table['inventory_turnover']>0]

invent_turnover_table.sort_values(by='dsi',ascending=False)
```

Now we are going to check how many products stay in the warehouse for more than one year, 180 days and 90 days respectevely. 

```python
dsi_365 = invent_turnover_table[invent_turnover_table['dsi']>365]
dsi_180 = invent_turnover_table[(invent_turnover_table['dsi']>180)&(invent_turnover_table['dsi']<=365)]
dsi_90 = invent_turnover_table[(invent_turnover_table['dsi']>90)&(invent_turnover_table['dsi']<=180)]
```

```python
print (f'There are {len(dsi_365)} products that take more than one year to be cleared from the inventory.')
print (f'There are {len(dsi_180)} products that take between one year and 180 days to be cleared from the inventory.')
print (f'There are {len(dsi_90)} products that take between 180 and 90 days to be cleared from the inventory.')
```

Since there are so many products that have a high days sales of inventory, we want to know if they have any caracteristics in common. So we check the price. We conclude that there are no similiarities between this products within each group. 

```python
print (dsi_365.price.mean())
print (dsi_365.price.max())
print (dsi_365.price.min())
print (np.median(dsi_365.price))
```

```python
print (dsi_180.price.mean())
print (dsi_180.price.max())
print (dsi_180.price.min())
print (np.median(dsi_180.price))
```

```python
print (dsi_90.price.mean())
print (dsi_90.price.max())
print (dsi_90.price.min())
print (np.median(dsi_90.price))
```

#### Monthly income

To know in which countries there could be a purchasing power that could be relevant for Zara we analyse the Gross National Income per country per capita.
We clean the data and drop all the NaN. 
The data we will be using is from 2017. Since usually there are not relevant changes in macroeconomy, the fact that this data is not very recent must not have a high impact. 
After this, we convert the currency to EUR (the currency in which our data is) and we divide it by 12, so we can have the data per month. 

After this we then multiply it by 5%, as this is the [recomended](https://budgeting.thenest.com/percent-income-budget-clothes-25211.html) percentage of the income to spend in apparel monthly.
